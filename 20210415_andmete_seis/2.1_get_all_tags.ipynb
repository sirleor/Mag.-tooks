```python
#resolutsioon
def resolutsioon(supp):
    resolutsioon = supp.findAll('p', class_=["C30Dispositifalinea", "C08Dispositif", "C31Dispositiftiretlong","C32Dispositifmarge1","C36Dispositifmarge2avectiretlong","C39Dispositifmarge1doubleretrait", "C37Dispositifmarge0doubleretrait", "C42Dispositifalinea","C02AlineaAltACharChar","C45Dispositifmarge1avectiretlong", "C44Dispositifmarge1avecretrait","C43Dispositiftiretlong","C33Dispositifmarge1avecretrait", "C40Dispositifmarge0avecretrait","C48DispositifIntroduction","C34Dispositifmarge1avectiretlong","C41DispositifIntroduction"])
    i=0
    while i < len(resolutsioon):
        resolutsioon[i]=resolutsioon[i].get_text().replace(u'\xa0', u' ')
        i+=1
    df_temp=pd.DataFrame(resolutsioon)
    df_temp.columns = ["Lõigu sisu"]
    df_temp['Pealkiri1'] = 'Resolutsioon' 
    return df_temp
    
resolutsioon(soup_from_txt)
```

```python
#pooled     
def poolte_nimed(supp):
    kogu_algus = supp.findAll('p', class_='C02AlineaAltA')
    poolte_nimed = []
    i=0
    while i != len(kogu_algus):
        if kogu_algus[i].findAll('b'):
            poolte_nimed.append('-'+kogu_algus[i].findAll('b')[0].get_text().replace(',', ''))
            i+=1
        else:
            i+=1
    kokku_nimed=' '.join(poolte_nimed)
    df_temp=pd.DataFrame(columns=['Lõigu sisu','Pealkiri1'], index=range(1))
    df_temp['Lõigu sisu'] = kokku_nimed
    df_temp['Pealkiri1'] = 'Pooled'
    return df_temp

poolte_nimed(soup_from_txt)
```

```python
# Kogu kohtulahendi kohta dataframe
def kohtulahend(supp):
    df_temp = pd.DataFrame(columns=['Lõigu number','Lõigu sisu','Pealkiri1','Alampealkiri2','Vahepealkiri3','Vahepealkiri4', 'Vahepealkiri5', 'Kohtuasja_nr', "Kohtuasja_ID"])
    count_C04Titre1=0
    for elem in soup_from_txt.findAll('p'): #arvutamaks välja palju peamisi pealkirju on
        if elem.has_attr('class'):
            if elem['class']==['C04Titre1']:
                count_C04Titre1+=1
    for i in range(count_C04Titre1):                                     #nii palju kui on pealkirju, siis nende järgi leiab osad ja liidab kokku
        if i==0:
            df_temp=df_temp.append(enne_oiguslikku(supp, i), ignore_index=True, sort=True)
            df_temp['Lõigu number'] = df_temp['Lõigu number']
            df_temp['Pealkiri1'] = 'Otsusest alates'
            df_temp.sort_values(by=['Lõigu number'])
        else:
            df_temp=df_temp.append(osad_frame(supp, i), ignore_index=True, sort=True)
            df_temp['Lõigu number'] = df_temp['Lõigu number']
            df_temp.sort_values(by=['Lõigu number'])
    df_temp=df_temp.append(koda(supp), ignore_index=True, sort=True)
    df_temp=df_temp.append(kuupaev(supp), ignore_index=True, sort=True)
    df_temp=df_temp.append(viited_marksonale(supp), ignore_index=True, sort=True)
    df_temp=df_temp.append(vahetekst_esemest(supp), ignore_index=True, sort=True)
    df_temp=df_temp.append(poolte_nimed(supp), ignore_index=True, sort=True)
#Kohtu koossseis on hetkel jäetud välja
    df_temp['Kohtuasja_nr']=kohtuasja_nr(supp)
    return df_temp#.iloc[1]['Lõigu sisu']

#df_temp.tail(20)
#df_temp.head(20)
#df_temp[df_temp['Pealkiri1']=='Pooled']
#df_temp[df_temp['Lõigu number']==9]
#row=[5]
#column=['Lõigu sisu']
#df_temp.loc[row, column]
#df_temp.iloc[4]['Lõigu sisu']
kohtulahend(soup_from_txt)
```
