```python
def query2soup(url):
    with urllib.request.urlopen(url) as response:
        html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
#Viited märksõnadele
def viited_marksonale(supp):
    viited = supp.findAll('p', class_='C71Indicateur')
    return viited[0].get_text().replace(u'\xa0', u' ')

viited_marksonale(soup_from_txt)
```

```python
#kuupäev
def kuupaev(supp):
    kuupaev = supp.findAll('p', class_="C19Centre")
    return kuupaev[1].get_text().replace('(*)', '').replace(u'\xa0', u' ')

kuupaev(soup_from_txt)
```

```python
#koda
def koda(supp):
    koda = supp.findAll('p', class_="C19Centre")
    return koda[0].get_text().replace(',','')

koda(soup_from_txt)
```

```python
#kohtuasja number
def kohtuasja_nr(supp):
    kohtuasja_nr = supp.find('p', class_='C02AlineaAltA')
    return kohtuasja_nr.get_text().replace(u'\xa0', u' ').replace(',','').replace('Kohtuasjas ', '')

kohtuasja_nr(soup_from_txt)
```

```python
#poolte nimede leidmine ja listi lisamine
def poolte_nimed(supp):
    kogu_algus = supp.findAll('p', class_='C02AlineaAltA')
    poolte_nimed = []
    i=0
    while i != len(kogu_algus):
        if kogu_algus[i].findAll('b'):
            poolte_nimed.append('-'+kogu_algus[i].findAll('b')[0].get_text())
            i+=1
        else:
            i+=1
    kokku_nimed=' '.join(poolte_nimed)
    return kokku_nimed

poolte_nimed(soup_from_txt)
```

```python
#resolutsioon
def resolutsioon(supp):
    resolutsioon = supp.findAll('p', class_="C30Dispositifalinea") or supp.findAll('p', class_="C08Dispositif")
    i=0
    while i < len(resolutsioon):
        resolutsioon[i]=resolutsioon[i].get_text().replace(u'\xa0', u' ')
        i+=1
    return resolutsioon
    
resolutsioon(soup_from_txt)
```

```python
# Esimene blokk andmeid Õiguslikust raamistikust kuni Põhikohtuasja ja eelotsuste küsimusteni

def esimene_blokk(supp):
    abilist=[]
    abilist2=[]
    count_C04Titre1=0
    count_C05Titre2=0
    for elem in supp.findAll('p'):
        if elem['class']==['C04Titre1']:
            count_C04Titre1+=1
        if elem['class']==['C05Titre2']:
            count_C05Titre2+=1
        if count_C04Titre1>0 and count_C04Titre1<2:
            if count_C05Titre2==1:
                lause=elem.get_text().replace(u'\xa0', u' ')  #element suppist
                abilist.append(lause)
            else:
                if count_C05Titre2>1 and count_C05Titre2<2:
                    lause=elem.get_text().replace(u'\xa0', u' ')  #element suppist
                    abilist.append(lause)
                else:
                    lause=elem.get_text().replace(u'\xa0', u' ')  #element suppist
                    abilist2.append(lause)
    if len(abilist2)>0:
        df_temp=pd.DataFrame(abilist[1:])                                      #dataframe loomine listi põhjal, kui on 2 alampealkirja
        df_temp=pd.DataFrame(df_temp[0].str.split('  ', 1).tolist())           
        df_temp['sektsioonityyp']='Õiguslik raamistik'
        df_temp['sektsioonialamtyyp']=abilist[0]
        df_temp.columns = ['Lõigu number', 'Lõigu sisu', 'Sektsioon', 'Alamtüüp']
        for index, row in df_temp.iterrows():
            if row['Lõigu sisu']==None:
                df_temp.loc[index, 'Lõigu sisu'] = row['Lõigu number']
                df_temp.loc[index, 'Lõigu number'] ='None'
            elif row['Lõigu number'].isdigit()==False and row['Lõigu number']!=None:
                df_temp.loc[index, 'Lõigu sisu'] = row['Lõigu number']+row['Lõigu sisu']
                df_temp.loc[index, 'Lõigu number'] ='None'
        df_temp2=pd.DataFrame(abilist2[2:])                                      #dataframe2 loomine listi põhjal
        df_temp2=pd.DataFrame(df_temp2[0].str.split('  ', 1).tolist())           
        df_temp2['sektsioonityyp']='Õiguslik raamistik'
        df_temp2['sektsioonialamtyyp']=abilist2[1]
        df_temp2.columns = ['Lõigu number', 'Lõigu sisu', 'Sektsioon', 'Alamtüüp']
        for index, row in df_temp2.iterrows():
            if row['Lõigu sisu']==None:
                df_temp2.loc[index, 'Lõigu sisu'] = row['Lõigu number']
                df_temp2.loc[index, 'Lõigu number'] ='None'
            elif row['Lõigu number'].isdigit()==False and row['Lõigu number']!=None:
                df_temp2.loc[index, 'Lõigu sisu'] = row['Lõigu number']+row['Lõigu sisu']
                df_temp2.loc[index, 'Lõigu number'] ='None'
        df_row_reindex = pd.concat([df_temp, df_temp2], ignore_index=True)
    else:
        df_temp=pd.DataFrame(abilist[1:])                                      #dataframe loomine listi põhjal, kui on 1 alampealkiri
        df_temp=pd.DataFrame(df_temp[0].str.split('  ', 1).tolist())           
        df_temp['sektsioonityyp']='Õiguslik raamistik'
        df_temp['sektsioonialamtyyp']=abilist[0]
        df_temp.columns = ['Lõigu number', 'Lõigu sisu', 'Sektsioon', 'Alamtüüp']
        for index, row in df_temp.iterrows():
            if row['Lõigu sisu']==None:
                df_temp.loc[index, 'Lõigu sisu'] = row['Lõigu number']
                df_temp.loc[index, 'Lõigu number'] ='None'
            elif row['Lõigu number'].isdigit()==False and row['Lõigu number']!=None:
                df_temp.loc[index, 'Lõigu sisu'] = row['Lõigu number']+row['Lõigu sisu']
                df_temp.loc[index, 'Lõigu number'] ='None'
        df_row_reindex=df_temp
    #Muudab need, kus on None Lõigu numbrid eelmiseks lõigu numbriks
    df_row_reindex['Lõigu number']=df_row_reindex['Lõigu number'].replace({'None':np.nan})
    df_row_reindex['Lõigu number']=df_row_reindex['Lõigu number'].fillna(method='ffill')
    esimene_blokk=df_row_reindex.groupby(['Lõigu number','Sektsioon', 'Alamtüüp'])['Lõigu sisu'].apply(lambda x: ' '.join(x)).reset_index() #Gruppeerib need read kokku, kus on sama Lõigu number
    esimene_blokk_o=esimene_blokk[['Lõigu number','Lõigu sisu', 'Sektsioon', 'Alamtüüp']] #paneb õigesse järjestusse columnid
    return esimene_blokk_o
            
esimene_blokk(soup_from_txt)
```

```python
# Teine blokk andmeid Põhikohtuasi ja eelotsuste küsimused kuni eelotsuste küsimuste analüüsini
def teine_blokk(supp):
    abilist=[]
    naeb_jutumarki=False
    count_C04Titre1=0     #C04Titre1 on pealkirjad, mida on kokku 4
    for elem in supp.findAll('p'):
        if elem['class']==['C04Titre1']:
            count_C04Titre1+=1
        if count_C04Titre1>1 and count_C04Titre1<3:   #kui pealkiri on vahemikus 2 ja ei ole veel 3, siis tee toiminguid
            lause=elem.get_text().replace(u'\xa0', u' ')  #element suppist
            if len(abilist)==0:
                abilist.append(lause)
            elif lause.startswith('„'):
                abilist[len(abilist)-1] = abilist[len(abilist)-1]+lause
                naeb_jutumarki=True
            elif lause.endswith('“'):
                abilist[len(abilist)-1] = abilist[len(abilist)-1]+lause
                naeb_jutumarki=False
            elif naeb_jutumarki:
                abilist[len(abilist)-1] = abilist[len(abilist)-1]+lause
            else:
                abilist.append(lause)
    df_temp=pd.DataFrame(abilist[1:])
    df_temp=pd.DataFrame(df_temp[0].str.split('  ', 1).tolist())
    df_temp['sektsioonityyp']='Põhikohtuasi'
    df_temp.columns = ['Lõigu number', 'Lõigu sisu', 'Sektsioon']
    return df_temp


teine_blokk(soup_from_txt)
```

```python
# Kolmas blokk andmeid kas kiirmenetlus või Eelotsuste küsimuste analüüs
def kolmas_blokk(supp):
    abilist=[]
    count_C04Titre1=0
    for elem in supp.findAll('p'):
        if elem['class']==['C04Titre1']:
            count_C04Titre1+=1
        if count_C04Titre1>2 and count_C04Titre1<4:
            abilist.append(elem.get_text().replace(u'\xa0', u' '))
    df_temp=pd.DataFrame(abilist[1:])                                      #dataframe loomine listi põhjal
    df_temp=pd.DataFrame(df_temp[0].str.split('  ', 1).tolist())           
    df_temp['sektsioonityyp']=abilist[0]
    df_temp.columns = ['Lõigu number', 'Lõigu sisu', 'Sektsioon']
    return df_temp

kolmas_blokk(soup_from_txt)
```
