```python
from bs4 import BeautifulSoup
import pandas as pd
from nltk import word_tokenize
```

```python
data = pd.read_csv("Pohitabel_filt_algusega.csv")
data.head()
```

```python
column_list=data['Lõigu sisu'].to_list()
```

```python
column_list_low = [str(x).lower() for x in column_list]  
```

```python
#Loome eesti stopsõnade listi, sh punktid jm kirjavahemärgid
# https://intellica-ai.medium.com/comparison-of-different-word-embeddings-on-text-similarity-a-use-case-in-nlp-e83e08469c1c
#https://datadoi.ee/handle/33/78
with open('estonian-stopwords.txt',encoding='utf-8') as stopid:
    stopid_s = stopid.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
stopid_s = [x.strip() for x in stopid_s] 
stopid_s
import string
stopset = stopid_s + list(string.punctuation)
```

```python
#eemaldame stop sõnad
clean_list=[]
for j in range(len(column_list_low)):
    corpus = word_tokenize(column_list_low[j])
    #print(corpus)
    x=[]
    for i in corpus:
        if i not in stopset:
            x.append(i)
    clean_list.append(x)
clean_list
```

```python
#Liidame üksikud sõnad uuesti kokku
stop_puhas=[' '.join(el) for el in clean_list]
stop_puhas
```

```python
#Lisame puhta tulba tabelisse
data['stop_eemaldatud'] = stop_puhas
```

```python
#Kirjutame uue tabeli eraldi faili, et ei peaks protsessi kordama
data.to_csv("Pohitabel_stopid_eemald_lower.csv", index=False)
```
